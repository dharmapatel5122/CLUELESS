```python
print("Choose category:\n" 
      "1. Agriculture\n" 
      "2. Household\n"
      "3. Industry\n"
      )
x = input("Enter category: ").lower()

if x == "agriculture":
    print("Choose category: fruits, vegetables, grains\n")
    y = input("Enter item name: ").capitalize()
    if y == "Fruits":
        print("Choose fruit: \n"
              "1. Mango\n"
              "2. Banana\n"
              "3. Apple\n"
              "4. Orange\n"
              "5. Grapes\n")
    elif y == "Vegetables":
        print("Choose vegetable: \n"
              "1. Tomatoes\n"
              "2. Potatoes\n"
              "3. Onions\n"
              "4. Carrots\n"
              "5. Cabbage\n")
    elif y == "Grains":
        print("Choose grain: \n"
              "1. Rice\n"
              "2. Wheat\n"
              "3. Corn\n"
              "4. Bajra\n"
              "5. Oats\n")
    exit()
elif x == "household":
    print("Please choose household item.\n"
        "1. cooking\n"
        "2. shower\n"
        "3. toilet flush\n"
        "4. laundry\n"
        "5. drinking\n"
        "6. cleaning\n"
         )
    exit()
elif x == "industry":
    print("Please choose industry item.\n"
        "1. textile industry\n"
        "2. pharmaceuticals industry\n"
        "3. construction industry\n"
        "4. paper industry\n"
        "5. hydropower\n"
        "6. steel industry\n"
         )
    exit()

# Define dictionaries for water footprint data
fruits = {
    "Mango": 900,    
    "Banana": 160,   
    "Apple": 560,    
    "Orange": 360,   
    "Grapes": 320   
}

vegetables = {
    "Tomatoes": 180,  
    "Potatoes": 100,  
    "Onions": 110,    # Water footprint in L/kg
    "Carrots": 130,   # Water footprint in L/kg
    "Cabbage": 240   # Water footprint in L/kg
}

grains = {
    "Rice": 2000,     # Water footprint in L/kg
    "Wheat": 900,     # Water footprint in L/kg
    "Corn": 900,      # Water footprint in L/kg
    "Bajra": 300,     # Water footprint in L/kg
    "Oats": 900      # Water footprint in L/kg
}

household_data = {
    "cooking": 5,
    "shower": 20,
    "toilet flush": 10,
    "laundry": 40,
    "drinking": 5,
    "cleaning": 50
}

industry_data = {
    "textile industry":200,
    "pharmaceuticals industry":1996,
    "construction industry":320,
    "paper industry":200,
    "hydropower":1892,
    "steel industry":300
}

def calculate_water_footprint(product, quantity):
    if x == "agriculture":
        if product in fruits:
            footprint_per_unit = fruits[product]
            total_footprint = footprint_per_unit * quantity
            return total_footprint
        elif product in vegetables:
            footprint_per_unit = vegetables[product]
            total_footprint = footprint_per_unit * quantity
            return total_footprint
        elif product in grains:
            footprint_per_unit = grains[product]
            total_footprint = footprint_per_unit * quantity
            return total_footprint
    elif x == "household":
        if product in household_data:
            footprint_per_unit = household_data[product]
            total_footprint1 = footprint_per_unit * quantity
            return total_footprint1
    elif x == "industry":
        if product in industry_data:
            footprint_per_unit = industry_data[product]
            total_footprint = footprint_per_unit * quantity
            return total_footprint

def main():
    while True:
        product = input("Enter a product name: ")
        if product.lower() == "quit":
            break

        if product not in fruits and product not in vegetables and product not in grains and product not in household_data and product not in industry_data:
            print("Product not found .")
            continue

        try:
            quantity = float(input("Enter the quantity consumed: "))
            if quantity < 0:
                print("Quantity cannot be negative.")
                continue

            total_footprint = calculate_water_footprint(product, quantity)
            
            if total_footprint is not None:
                if x == "household":
                    print(f"Water footprint for {quantity} person of {product}: {total_footprint}  liters")
                else:
                    print(f"Water footprint for {quantity} kg of {product}: {total_footprint}  liters")
            else:
                print("Product not found in the database.")
        except ValueError:
            print("Invalid input. Please enter a valid quantity.")

if __name__ == "__main__":
    main()
```
